# Parallel LLM testing notebook
This is a notebook for a side-by-side comparison of SBERT models.


```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
# test pair to check each model
test_pair = {
    'source': "يمكن لامرأة قوية تجد الحب؟",  # our example from before
    'target': "Can a strong woman find love?",
    'language': 'ar'
}

# models to try (these are supposed to be good with different languages)
models = [
    'sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2',
    'sentence-transformers/LaBSE',
    'sentence-transformers/multi-qa-mpnet-base-dot-v1',
    'sentence-transformers/all-MiniLM-L12-v2',
    'sentence-transformers/all-distilroberta-v1'
]

# test each model
print("Testing different models...")
for model_name in models:
    print(f"\nTrying: {model_name}")
    try:
        # load the model
        model = SentenceTransformer(model_name)
        
        # get embeddings
        source_emb = model.encode(test_pair['source'])
        target_emb = model.encode(test_pair['target'])
        
        # calculate similarity
        sim = np.dot(source_emb, target_emb) / (np.linalg.norm(source_emb) * np.linalg.norm(target_emb))
        
        print(f"Source ({test_pair['language']}): {test_pair['source']}")
        print(f"Target (en): {test_pair['target']}")
        print(f"Similarity: {sim:.3f}")
        
        # quick English check
        eng_test = ["What is your name?", "What's your name?"]
        eng_emb1 = model.encode(eng_test[0])
        eng_emb2 = model.encode(eng_test[1])
        eng_sim = np.dot(eng_emb1, eng_emb2) / (np.linalg.norm(eng_emb1) * np.linalg.norm(eng_emb2))
        print(f"English control similarity: {eng_sim:.3f}")
        
    except Exception as e:
        print(f"Oops, something went wrong with this model: {str(e)}")

print("\nDone testing!")
```
